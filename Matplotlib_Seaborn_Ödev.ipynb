```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips
```

Grafiklerin her birinde xlabel, ylabel ve title olsun. Başkası grafiğe baktığında anlayabilsin, grafikler kendini açıklayıcı olsun.

Renklendirmeler ve marker seçimleri uyumlu olsun.

Parametreler için seaborn'un kendi sitesinden faydalanabilirsiniz.

```python
# Soru: Sigara içen ve içmeyen müşterilerin fatura tutarlarının dağılımını gösteren bir histogram oluşturun.
```

```python
sns.histplot(data = tips, x = "total_bill", hue = "smoker", alpha= 0.3)
plt.title("Sigara İçen ve İçmeyenlerin Fatura Tutar Dağılımı")
plt.xlabel("Total Bill")
plt.ylabel("Smoker Count")
plt.show()
```

```python
# Soru: Veri setindeki "total_bill" değişkeninin dağılımını gösteren bir histogram oluşturun.
```

```python
sns.histplot(data = tips, x = "total_bill")
plt.title("Fatura Tutar Dağılımı")
plt.xlabel("Total Bill")
plt.show()
```

```python
# Soru: Veri setindeki "total_bill" ve "tip" değişkenleri arasındaki ilişkiyi gösteren bir 
# scatterplot oluşturun. Ayrıca, "time" değişkenini kullanarak noktaları renklendirin.
```

```python
sns.scatterplot(
    data=tips,
    x = "total_bill",
    y = "tip",
    hue = "time"
)

plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# Soru: Veri setindeki "day" ve "total_bill" değişkenlerini kullanarak, 
# her gün için ortalama toplam fatura tutarını gösteren bir barplot oluşturun.
```

```python
mean_total_bill = tips.groupby(["day"])["total_bill"].mean()
mean_total_bill # soruda isteneni verecek şekilde bir data oluşturduk.
```

```python
mean_total_bill = tips.groupby(["day"])["total_bill"].mean().reset_index() #indexlerini sıfırladık.
sns.barplot(
    data = mean_total_bill,
    x ="day",
    y = "total_bill"
)

plt.xlabel("Days")
plt.ylabel("Average Total Bill ")
plt.title("Average Total Bill for Days")
plt.show()
```

```python
# Soru: "tips" veri setinde, her günün bahşiş oranını (yüzde olarak) hesaplayın 
# ve bunları gösteren bir line plot oluşturun.
```

```python
per_tip = tips["tip"] * 100
per_tip.index = tips["day"]
per_tip
```

```python
sns.lineplot(
    data = per_tip,
    x = per_tip.index,
    y = per_tip.values,
)

plt.title("Günlere Göre Ortalama Bahşiş")
plt.xlabel("Day")
plt.ylabel("Percentege(%)")
plt.show()
```

```python
# Soru: "tips" veri setinde, her gün için fatura tutarının ve bahşiş miktarının ortalamasını hesaplayarak, 
# iki değişken arasındaki ilişkiyi gösteren bir scatterplot oluşturun.
```

```python
mean = tips.groupby(["day"])["total_bill", "tip"].mean()
mean
```

```python
sns.scatterplot(
    data = mean,
    x = "total_bill",
    y = "tip",
    hue = "day"
)

plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# Soru: Sigara içen ve içmeyen müşterilerin, farklı günlerde ödedikleri toplam
# bahşiş miktarını gösteren bir bar plot oluşturun.
```

```python
plt.figure(figsize=(12,8))

sns.barplot(
    data = tips,
    x = "day",
    y = "total_bill",
    hue = "smoker"
)

plt.title("Day & Total Bill")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Soru: Sigara içen ve içmeyen müşterilerin ödedikleri bahşiş miktarlarının
# dağılımını gösteren bir boxplot oluşturun.
```

```python
sns.boxplot(
    data = tips,
    x = "smoker",
    y = "tip",
    hue = "smoker"
)
plt.title("Smokers vs Tip")
plt.xlabel("Smoker")
plt.ylabel("Tip")
plt.show()
```

```python
# Soru: Yukarıdan seçeceğiniz 4 farklı plotu, subplot kullanarak çizdirin.
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,10))

sns.histplot(data = tips, 
             x = "total_bill", 
             hue = "smoker", 
             alpha= 0.3,
             ax = axes[0,0]
)
axes[0,0].set_title("Sigara İçen ve İçmeyenlerin Fatura Tutar Dağılımı")


sns.scatterplot(
    data=tips,
    x = "total_bill",
    y = "tip",
    hue = "time",
    ax = axes[0,1]
)
axes[0,1].set_title("Total Bill vs Tip")

plt.subplots_adjust(wspace=0.5, hspace=0.5)


sns.barplot(
    data = mean_total_bill,
    x ="day",
    y = "total_bill",
    ax = axes[1,0]
)
axes[0,1].set_title("Average Total Bill for Days")


sns.barplot(
    data = tips,
    x = "day",
    y = "total_bill",
    hue = "smoker",
    ax = axes[1,1]
)
axes[1,1].set_title("Day & Total Bill")
```
