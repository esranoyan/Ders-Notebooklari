```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips
```

Grafiklerin her birinde xlabel, ylabel ve title olsun. BaÅŸkasÄ± grafiÄŸe baktÄ±ÄŸÄ±nda anlayabilsin, grafikler kendini aÃ§Ä±klayÄ±cÄ± olsun.

Renklendirmeler ve marker seÃ§imleri uyumlu olsun.

Parametreler iÃ§in seaborn'un kendi sitesinden faydalanabilirsiniz.

```python
# Soru: Sigara iÃ§en ve iÃ§meyen mÃ¼ÅŸterilerin fatura tutarlarÄ±nÄ±n daÄŸÄ±lÄ±mÄ±nÄ± gÃ¶steren bir histogram oluÅŸturun.
```

```python
sns.histplot(data = tips, x = "total_bill", hue = "smoker", alpha= 0.3)
plt.title("Sigara Ä°Ã§en ve Ä°Ã§meyenlerin Fatura Tutar DaÄŸÄ±lÄ±mÄ±")
plt.xlabel("Total Bill")
plt.ylabel("Smoker Count")
plt.show()
```

```python
# Soru: Veri setindeki "total_bill" deÄŸiÅŸkeninin daÄŸÄ±lÄ±mÄ±nÄ± gÃ¶steren bir histogram oluÅŸturun.
```

```python
sns.histplot(data = tips, x = "total_bill")
plt.title("Fatura Tutar DaÄŸÄ±lÄ±mÄ±")
plt.xlabel("Total Bill")
plt.show()
```

```python
# Soru: Veri setindeki "total_bill" ve "tip" deÄŸiÅŸkenleri arasÄ±ndaki iliÅŸkiyi gÃ¶steren bir 
# scatterplot oluÅŸturun. AyrÄ±ca, "time" deÄŸiÅŸkenini kullanarak noktalarÄ± renklendirin.
```

```python
sns.scatterplot(
    data=tips,
    x = "total_bill",
    y = "tip",
    hue = "time"
)

plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# Soru: Veri setindeki "day" ve "total_bill" deÄŸiÅŸkenlerini kullanarak, 
# her gÃ¼n iÃ§in ortalama toplam fatura tutarÄ±nÄ± gÃ¶steren bir barplot oluÅŸturun.
```

```python
mean_total_bill = tips.groupby(["day"])["total_bill"].mean()
mean_total_bill # soruda isteneni verecek ÅŸekilde bir data oluÅŸturduk.
```

```python
mean_total_bill = tips.groupby(["day"])["total_bill"].mean().reset_index() #indexlerini sÄ±fÄ±rladÄ±k.
sns.barplot(
    data = mean_total_bill,
    x ="day",
    y = "total_bill"
)

plt.xlabel("Days")
plt.ylabel("Average Total Bill ")
plt.title("Average Total Bill for Days")
plt.show()
```

```python
# Soru: "tips" veri setinde, her gÃ¼nÃ¼n bahÅŸiÅŸ oranÄ±nÄ± (yÃ¼zde olarak) hesaplayÄ±n 
# ve bunlarÄ± gÃ¶steren bir line plot oluÅŸturun.
```

```python
per_tip = tips["tip"] * 100
per_tip.index = tips["day"]
per_tip
```

```python
sns.lineplot(
    data = per_tip,
    x = per_tip.index,
    y = per_tip.values,
)

plt.title("GÃ¼nlere GÃ¶re Ortalama BahÅŸiÅŸ")
plt.xlabel("Day")
plt.ylabel("Percentege(%)")
plt.show()
```

```python
# Soru: "tips" veri setinde, her gÃ¼n iÃ§in fatura tutarÄ±nÄ±n ve bahÅŸiÅŸ miktarÄ±nÄ±n ortalamasÄ±nÄ± hesaplayarak, 
# iki deÄŸiÅŸken arasÄ±ndaki iliÅŸkiyi gÃ¶steren bir scatterplot oluÅŸturun.
```

```python
mean = tips.groupby(["day"])["total_bill", "tip"].mean()
mean
```

```python
sns.scatterplot(
    data = mean,
    x = "total_bill",
    y = "tip",
    hue = "day"
)

plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# Soru: Sigara iÃ§en ve iÃ§meyen mÃ¼ÅŸterilerin, farklÄ± gÃ¼nlerde Ã¶dedikleri toplam
# bahÅŸiÅŸ miktarÄ±nÄ± gÃ¶steren bir bar plot oluÅŸturun.
```

```python
plt.figure(figsize=(12,8))

sns.barplot(
    data = tips,
    x = "day",
    y = "total_bill",
    hue = "smoker"
)

plt.title("Day & Total Bill")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Soru: Sigara iÃ§en ve iÃ§meyen mÃ¼ÅŸterilerin Ã¶dedikleri bahÅŸiÅŸ miktarlarÄ±nÄ±n
# daÄŸÄ±lÄ±mÄ±nÄ± gÃ¶steren bir boxplot oluÅŸturun.
```

```python
sns.boxplot(
    data = tips,
    x = "smoker",
    y = "tip",
    hue = "smoker"
)
plt.title("Smokers vs Tip")
plt.xlabel("Smoker")
plt.ylabel("Tip")
plt.show()
```

```python
# Soru: YukarÄ±dan seÃ§eceÄŸiniz 4 farklÄ± plotu, subplot kullanarak Ã§izdirin.
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,10))

sns.histplot(data = tips, 
             x = "total_bill", 
             hue = "smoker", 
             alpha= 0.3,
             ax = axes[0,0]
)
axes[0,0].set_title("Sigara Ä°Ã§en ve Ä°Ã§meyenlerin Fatura Tutar DaÄŸÄ±lÄ±mÄ±")


sns.scatterplot(
    data=tips,
    x = "total_bill",
    y = "tip",
    hue = "time",
    ax = axes[0,1]
)
axes[0,1].set_title("Total Bill vs Tip")

plt.subplots_adjust(wspace=0.5, hspace=0.5)


sns.barplot(
    data = mean_total_bill,
    x ="day",
    y = "total_bill",
    ax = axes[1,0]
)
axes[0,1].set_title("Average Total Bill for Days")


sns.barplot(
    data = tips,
    x = "day",
    y = "total_bill",
    hue = "smoker",
    ax = axes[1,1]
)
axes[1,1].set_title("Day & Total Bill")
```
