   Not : Eğer bir öğrencinin sadece ikinci dönem notu bulunuyorsa, birinci dönem hesaba eklenmeden ikinci dönem notu yıl sonu notu olarak hesaplansın.

```python
not2_data = [ 80, 92 , 85 , 87, 90 , 92, 100, 83]
new_isim_index = ["Ali" ,"Deniz" ,"Aslı" ,"Burak" ,"Kerem" ,"Yasemin" ,"Derya" ,"Mert"]

donem2_mat = pd.Series(not2_data, new_isim_index)
donem2_mat
```

```python
ort = donem1_mat.add(donem2_mat, fill_value=83) / 2
ort
```
